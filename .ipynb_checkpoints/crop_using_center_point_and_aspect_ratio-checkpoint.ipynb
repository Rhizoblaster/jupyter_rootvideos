```python
from PIL import Image

def crop_image(image_path, aspect_ratio, zoom, center):
    # Open the image
    image = Image.open(image_path)
    
    # Get the dimensions of the image
    width, height = image.size
    
    # Calculate the size of the crop
    crop_width = int(min(width, height * aspect_ratio) * zoom)
    crop_height = int(min(height, width / aspect_ratio) * zoom)
    
    # Calculate the left, upper, right, and lower coordinates of the crop
    left = int(center[0] * width - crop_width / 2)
    upper = int(center[1] * height - crop_height / 2)
    right = left + crop_width
    lower = upper + crop_height
    
    # Crop the image
    cropped_image = image.crop((left, upper, right, lower))
    
    # Return the cropped image
    return cropped_image
```
